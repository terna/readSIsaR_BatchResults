```python
import pandas as pd
import numpy as np
rundf = pd.read_csv('results/SIsaR_0.9.4.2 experiments 2 seeds-table_1000_NH_OP_BZat20_NOschoolSept.csv',
                    sep=',',skiprows=6)
```

```python
rundf["totalInfected"]=rundf["cumulativeRedCount"]+rundf["cumulativeVioletCount"]
rundf["totalInfected&Deceased"]=rundf["totalInfected"]+rundf["cumulativeDeadCount"]
rundf["totalInfected&Deceased"]=rundf["totalInfected"]+rundf["cumulativeDeadCount"]
```

```python
rundf.columns
```

```python
rundf.replace(-1, np.nan, inplace=True)
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

# Heatmaps

```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors

#import numpy as np
import seaborn as sns; sns.set()
```

### bins = 10

### symptomatic + asymptomatic + deceases

```python
hist = plt.hist2d(x=rundf['ticks'], y=rundf['totalInfected&Deceased'], bins=10,
                  range=[[0, 800], [0, 3600]])
```

```python
plt.figure(figsize=(8.5,4.25))
ax = sns.heatmap(
    np.flip(np.transpose(hist[0]),0), 
    robust=True, 
    cmap='YlGnBu',
    annot=True, fmt='.0f', 
    xticklabels=np.round(hist[1][1:],0).astype(int),
    yticklabels=np.flip(hist[2][1:],0).astype(int),
    annot_kws={"size":8}
)

plt.xlabel("Days")
plt.ylabel("Total symptomatic + asymptomatic + deceased")

plt.title('1000 epidemics with the absolut isolation of NHs at day 20,\nnon-pharmaceutical containment measures,\nno school in Sep')

plt.show()
```
